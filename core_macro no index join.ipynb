```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
%matplotlib inline
```

```python
# import and sort tables
file = Path("/content/drive/MyDrive/Bootcamp/Project_2/Data/Data tables/all_cusips_core_rating_shifted_df_v4.csv")
file2 = Path("/content/drive/MyDrive/Bootcamp/Project_2/Data/Data tables/cleansed_fred_macro_daily.csv")

df_core = pd.read_csv(file, infer_datetime_format=True, parse_dates=True)
df_macro = pd.read_csv(file2, infer_datetime_format=True, parse_dates=True)

df_core['date_fundamentals'] = pd.to_datetime(df_core['date_fundamentals'])
df_macro['observed_date'] = pd.to_datetime(df_macro['observed_date'])

df_core.sort_values('date_fundamentals', inplace=True)
df_macro.sort_values('observed_date', inplace=True)

print(df_core.info)
df_core.head()
```

```python
print(df_macro.info)
df_macro.head()
```

```python
# select features
df_macro_red = df_macro[[
                            'observed_date',
                            'percent_fred_macro_yc_30_yr',
                            'percent_fred_macro_yc_3_mo',
                            'percentchange_fred_macro_gdp',
                            'dollars_fred_crude_oil_prices',
                            'percent_fred_five_year_breakeven_inflation',
                            'index_fred_macro_vix',
                            ]]

# macro drop Nans. Count before: 6,308. Count after:  
df_macro_red.dropna(inplace=True)
print(df_macro_red.info)
df_macro_red.tail()
```

```python
# merge core and macro dataframes
tol = pd.Timedelta('3 day')
df3 = pd.merge_asof(left=df_core,right=df_macro_red, left_on='date_fundamentals', right_on='observed_date', direction='nearest',tolerance=tol)
print(df3.info)
df3.tail(10)
```

```python
output = Path("/content/drive/MyDrive/Bootcamp/Project_2/Data/Data tables/all_cusips_core_rating_shifted_df_v4_MACRO.csv")
df3.to_csv(output)
```
